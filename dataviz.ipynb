```python
import pandas as pd

# Charger les données
data = pd.read_csv('results.csv')
```

```python
# Afficher un aperçu des données
print(data.head())

# Convertir les colonnes nécessaires en types appropriés
data['run_date'] = pd.to_datetime(data['run_date'], format='%Y-%m-%d_%H-%M-%S')
data['actual_run'] = data['actual_run'].astype(int)
data['time'] = data['time'].astype(int)
data['score'] = data['score'].astype(int)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Créer un graphique pour chaque environnement
environments = data['env'].unique()

for env in environments:
    subset = data[data['env'] == env]

    fig, ax1 = plt.subplots(figsize=(12, 8))

    # Tracer le score
    sns.barplot(x='model', y='score', hue='actual_run', data=subset, ci=None, ax=ax1)
    ax1.set_title(f'Performance des modèles dans l\'environnement {env}')
    ax1.set_xlabel('Modèle')
    ax1.set_ylabel('Score')
    ax1.legend(title='Actual Run', loc='upper right')
    ax1.tick_params(axis='x', rotation=45)

    # Ajouter un deuxième axe pour le temps
    ax2 = ax1.twinx()
    sns.lineplot(x='model', y='time', hue='actual_run', data=subset, marker='o', ax=ax2, legend=False)
    ax2.set_ylabel('Temps (ms)')

    plt.tight_layout()
    plt.show()
```
