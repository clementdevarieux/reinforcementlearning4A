## Première approche

```python
#               bot          !          
#            P   F   C       !
#        P   0  -1   1       !
# nous   F   1   0  -1       !
#        C  -1   1   0       !
#                            !
#      matrice pour nous     !

S = [0, 1, 2] # pierre, feuille, ciseaux
A = [0, 1, 2] # pierre, feuille, ciseaux
R = [-2, -1, 0, 1, 2] # reward potentiel à la fin des deux manches
T = [0, 1, 2] # on finit sur pierre feuille ou ciseaux

p = np.zeros((len(S), len(A), len(S), len(R))) # S, A, S_p, R
```

```python
for s in range(len(S)):
    for a in range(len(A)):
        for s_p in range(len(S)):
            for r in range(len(R)):
                # on a joué pierre tous les deux , donc j'ai un reward de 0
                # il joue donc pierre le coup suivant
                # je gagne ensuite donc je joue feuille
                if s == 0 and a == 1 and s_p == 1 and R[r] == 1:
                    p[s][a][s_p][r] = 1.0
                # je perds ensuite, donc je joue ciseaux
                if s == 0 and a == 2 and s_p == 2 and R[r] == -1:
                    p[s][a][s_p][r] = 1.0
                # je draw ensuite donc je joue pierre encore
                if s == 0 and a == 0 and s_p == 0 and R[r] == 0:
                    p[s][a][s_p][r] = 1.0
                    
                # j'ai joué pierre, il a joué feuille, donc j'ai perdu la premiere manche
                # il joue donc pierre au tour suivant
                # je gagne ensuite donc je joue feuille
                if s == 0 and a == 1 and s_p == 1 and R[r] == 0:
                    p[s][a][s_p][r] = 1.0
                # je perds ensuite, donc je joue ciseaux
                if s == 0 and a == 2 and s_p == 2 and R[r] == -2:
                    p[s][a][s_p][r] = 1.0
                # je draw ensuite, donc je joue pierre
                if s == 0 and a == 0 and s_p == 0 and R[r] == -1:
                    p[s][a][s_p][r] = 1.0
                    
                # j'ai joué pierre et il a joué ciseaux, donc j'ai gagné la première manche
                # il joue donc pierre au tour suivant
                # je gagne la 2em manche, donc je joue feuille
                if s == 0 and a == 1 and s_p == 1 and R[r] == 2:
                    p[s][a][s_p][r] = 1.0
                # je perds la 2em manche, donc je joue ciseaux
                if s == 0 and a == 2 and s_p == 2 and R[r] == 0:
                    p[s][a][s_p][r] = 1.0
                # je draw la 2em manche, donc je joue pierre
                if s == 0 and a == 0 and s_p == 0 and R[r] == 1:
                    p[s][a][s_p][r] = 1.0
                    
        
                # on a joué feuille tous les deux , donc j'ai un reward de 0
                # il joue donc feuille le coup suivant
                # je gagne ensuite donc je joue ciseaux
                if s == 1 and a == 2 and s_p == 2 and R[r] == 1:
                    p[s][a][s_p][r] = 1.0
                # je perds ensuite, donc je joue pierre
                if s == 1 and a == 0 and s_p == 0 and R[r] == -1:
                    p[s][a][s_p][r] = 1.0
                # je draw ensuite donc je joue feuille encore
                if s == 1 and a == 1 and s_p == 1 and R[r] == 0:
                    p[s][a][s_p][r] = 1.0
                    
                # j'ai joué feuille, il a joué ciseaux, donc j'ai perdu la premiere manche
                # il joue donc feuille au tour suivant
                # je gagne ensuite donc je joue ciseaux
                if s == 1 and a == 2 and s_p == 2 and R[r] == 0:
                    p[s][a][s_p][r] = 1.0
                # je perds ensuite, donc je joue pierre
                if s == 1 and a == 0 and s_p == 0 and R[r] == -2:
                    p[s][a][s_p][r] = 1.0
                # je draw ensuite, donc je joue feuille
                if s == 1 and a == 1 and s_p == 1 and R[r] == -1:
                    p[s][a][s_p][r] = 1.0
                    
                # j'ai joué feuille et il a joué pierre, donc j'ai gagné la première manche
                # il joue donc feuille au tour suivant
                # je gagne la 2em manche, donc je joue ciseaux
                if s == 1 and a == 2 and s_p == 2 and R[r] == 2:
                    p[s][a][s_p][r] = 1.0
                # je perds la 2em manche, donc je joue pierre
                if s == 1 and a == 0 and s_p == 0 and R[r] == 0:
                    p[s][a][s_p][r] = 1.0
                # je draw la 2em manche, donc je joue feuille
                if s == 1 and a == 1 and s_p == 1 and R[r] == 1:
                    p[s][a][s_p][r] = 1.0
                    
                # ----------------------------
                #S = [0, 1, 2] # pierre, feuille, ciseaux
                #A = [0, 1, 2] # pierre, feuille, ciseaux
                #R = [-2, -1, 0, 1, 2] 
                
                # on a joué ciseaux tous les deux , donc j'ai un reward de 0
                # il joue donc ciseaux le coup suivant
                # je gagne ensuite donc je joue pierre
                if s == 2 and a == 0 and s_p == 0 and R[r] == 1:
                    p[s][a][s_p][r] = 1.0
                # je perds ensuite, donc je joue feuille
                if s == 2 and a == 1 and s_p == 1 and R[r] == -1:
                    p[s][a][s_p][r] = 1.0
                # je draw ensuite donc je joue ciseaux encore
                if s == 2 and a == 2 and s_p == 2 and R[r] == 0:
                    p[s][a][s_p][r] = 1.0
                    
                # j'ai joué ciseaux, il a joué pierre, donc j'ai perdu la premiere manche
                # il joue donc ciseaux au tour suivant
                # je gagne ensuite donc je joue pierre
                if s == 2 and a == 0 and s_p == 0 and R[r] == 0:
                    p[s][a][s_p][r] = 1.0
                # je perds ensuite, donc je joue feuille
                if s == 2 and a == 1 and s_p == 1 and R[r] == -2:
                    p[s][a][s_p][r] = 1.0
                # je draw ensuite, donc je joue ciseaux
                if s == 2 and a == 2 and s_p == 2 and R[r] == -1:
                    p[s][a][s_p][r] = 1.0
                    
                # j'ai joué ciseaux et il a joué feuille, donc j'ai gagné la première manche
                # il joue donc ciseaux au tour suivant
                # je gagne la 2em manche, donc je joue pierre
                if s == 2 and a == 0 and s_p == 0 and R[r] == 2:
                    p[s][a][s_p][r] = 1.0
                # je perds la 2em manche, donc je joue feuille
                if s == 2 and a == 1 and s_p == 1 and R[r] == 0:
                    p[s][a][s_p][r] = 1.0
                # je draw la 2em manche, donc je joue ciseaux
                if s == 2 and a == 2 and s_p == 2 and R[r] == 1:
                    p[s][a][s_p][r] = 1.0
```

```python
def naive_policy_iteration():
  theta = 0.00001
  gamma = 0.999

  V = np.random.random(len(S))
  for s in T:
    V[s] = 0.0
  Pi = np.random.choice(A, len(S), True)

  while True:
    # Policy evaluation
    while True:
      delta = 0
      for s in range(len(S)):
        v = V[s]
        total = 0
        for s_p in range(len(S)):
          for r in range(len(R)):
            total += p[s, Pi[s], s_p, r] * (R[r] + 0.999 * V[s_p])
        V[s] = total
        delta = np.maximum(delta, np.abs(v - V[s]))
      if delta < theta:
        break

    # Policy improvement
    policy_stable = True

    for s in S:
      if s in T:
        continue

      old_action = Pi[s]

      # Compute Arg Max a
      argmax_a = None
      max_a = -999999999

      for a in A:
        total = 0.0

        for s_p in S:
          for r_index in range(len(R)):
            total += p[s, a, s_p, r_index] * (R[r_index] + gamma * V[s_p])

        if argmax_a is None or total >= max_a:
          argmax_a = a
          max_a = total

      Pi[s] = argmax_a

      if old_action != Pi[s]:
        policy_stable = False


    if policy_stable:
      break

  return Pi
```
